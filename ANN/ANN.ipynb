# Part 1 - Data Preprocessing

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
X #CreditScore, ...
```

#### 1. Encoding categorical data

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
#encode country
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
X 
```

```python
#encode gender
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
X = pd.DataFrame(X)
X
```

##### Convert country category into one-hot encoding
Country becomes three columns -> binary mode

```python
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray() #Column: 0-2
X = X[:, 1:] #Exclude the col to avoid dummy variable trap (one variable can be predicted from others/variable redundant)
```

```python
X
```

#### 2. Splitting the dataset into the Training set and Test set

```python
X=pd.DataFrame(X)
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

#### 3. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train = pd.DataFrame(X_train)
```

# Part 2 - Make the ANN

```python
import keras
from keras.models import Sequential #initialize NN
from keras.layers import Dense #create layer
```

### 1. Initialising the ANN

```python
classifier = Sequential()
```

### 2. Apply the activation function

#### adding the input layer and first hidden layer

1. output_dim: the number of nodes of the layer we are adding in this function
= the average of the number of nodes in input layer (12, 12 attributes) and the number of nodes in output layer (1)
= 6

2. init: initialize the weights to a small number close to zero
    uniform: initialize the function according to a uniform distribution

3. activation function

4. input_dim: independent variables (only need to be created when creating the first hidden layer)

```python
classifier.add(Dense(output_dim=6, init='uniform', activation='relu', input_dim=11))
```

#### adding the second hidden layer

```python
classifier.add(Dense(output_dim=6, init='uniform', activation='relu'))
```

#### adding the output layer

```python
classifier.add(Dense(output_dim=1, init='uniform', activation='sigmoid'))
```

#### Compiling the ANN

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

#### Fitting the ANN to the Training set
1. batch_size:  Number of samples per gradient update
2. epoch: An epoch is an iteration over the entire x and y data provided.

```python
classifier.fit(X_train, y_train, batch_size=10, nb_epoch=100)
```

# Part 3 - Making the predictions and evaluating the model

```python
y_pred = classifier.predict(X_test)
y_pred
```

#### Convert the probability to true or false

```python
y_pred = (y_pred > 0.5) #if > 0.5, return true
y_pred
```

#### Making the confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

## Part 4 - Evaluating, Improving and Tuning the ANN
### Evaluating the ANN

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(output_dim=6, init='uniform', activation='relu', input_dim=11))
    classifier.add(Dense(output_dim=6, init='uniform', activation='relu'))
    classifier.add(Dense(output_dim=1, init='uniform', activation='sigmoid'))
    classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return classifier
```

```python
classifier = KerasClassifier(build_fn=build_classifier, batch_size=10, epochs=100)
# build_fn : a keras model
```

```python
accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10, n_jobs=1) 
#average accuracy from 10-fold cv
#n_jobs: -1 to use all the CPUs to train the model
```

```python
mean = accuracies.mean()
variance = accuracies.std()
variance
```

### Dropout regulazation to reduce overfitting

#### Tuning the ANN

```python
from keras.layers import Dropout
from sklearn.model_selection import GridSearchCV
```

when you have overfitting, you should apply dropout to all the layers

```python
def build_classifier_withDropout(optimizer, init):
    classifier = Sequential()
    classifier.add(Dense(output_dim=6, init=init, activation='relu', input_dim=11))
    #classifier.add(Dropout(p=0.1))
    #Dropout(p, )
    #p: 0-1, fraction of the neurons to drop (disabled).
    #   recommendation: start from 0.1, until overfitting solved. Not exceed 0.5, easy to cause underfitting
    
    classifier.add(Dense(output_dim=6, init=init, activation='relu'))
    #classifier.add(Dropout(p=0.1))
    
    classifier.add(Dense(output_dim=1, init=init, activation='sigmoid'))
    classifier.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return classifier
tuned_classifier = KerasClassifier(build_fn=build_classifier_withDropout)

parameters = {'batch_size': [5, 25, 32],
             'epochs': [100, 500],
             'optimizer': ['adam', 'rmsprop'],
             'init': ['uniform', 'glorot_uniform']}
#comes from experience

grid_search = GridSearchCV(estimator=tuned_classifier, 
                          param_grid = parameters,
                          scoring = 'accuracy',
                          cv = 10)
grid_search = grid_search.fit(X_train, y_train)
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
best_accuracy
```

accuracy: 0.8506
best parameters: batch_size=25, nb_epoch=500
